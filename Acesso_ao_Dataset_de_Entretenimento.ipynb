```python
import pandas as pd
import matplotlib.pyplot as plt

# DESAFIO PRÁTICO

# CÓDIGO PYTHON:

# Carregando o dataset a partir do arquivo CSV local
file_path = "netflix_titles.csv"
data = pd.read_csv(file_path)

# Exibindo as primeiras linhas do dataframe
print(data.head())

# 1. Quais colunas estão presentes no dataset?
# Obtendo a lista de colunas do dataframe
colunas = data.columns.tolist()
print("Colunas presentes no dataset:", colunas)

# 2. Quantos filmes estão disponíveis na Netflix?
# Contando o número de entradas onde o tipo é 'Movie'
num_filmes = data[data['type'] == 'Movie'].shape[0]
print("Número de filmes disponíveis na Netflix:", num_filmes)

# Analisando o número de produções lançadas por ano
producoes_por_ano = data['release_year'].value_counts()
print("Produções por ano:")
display(producoes_por_ano.sort_index())

# Gráfico 1: Produções por ano
plt.figure(figsize=(12, 6))
producoes_por_ano.sort_index().plot(kind='bar', color='skyblue')
plt.title('Número de Produções Lançadas por Ano')
plt.xlabel('Ano')
plt.ylabel('Número de Produções')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Analisando o número de produções por gênero
# Separando os gêneros e contando as ocorrências
categorias = data['listed_in'].str.split(', ').explode().value_counts()
print("Número de produções por gênero:")
display(categorias)

# Gráfico 2: Produções por gênero
plt.figure(figsize=(12, 6))
categorias.head(10).plot(kind='bar', color='lightgreen')
plt.title('Número de Produções por Gênero (Top 10)')
plt.xlabel('Gênero')
plt.ylabel('Número de Produções')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Diretores prolíficos
# Contando o número de produções para os 5 diretores mais ativos
diretores_count = data['director'].value_counts().head(5)
print("5 diretores com mais filmes e séries na plataforma:")
display(diretores_count)

# 4. Diretores que também atuaram como atores
# Verificando quais diretores estão na lista de elenco (cast) de suas produções
diretores_atores = data[data['director'].isin(data['cast'].str.split(', ').explode())]['director'].unique()
print("Diretores que também atuaram como atores em suas produções:")
display(diretores_atores)

# 5. Duração média dos filmes
# Se a coluna 'duration' estiver presente, calcular a duração média
if 'duration' in data.columns:
    # Removendo o texto ' min' para converter a coluna em numérica
    data['duration'] = data['duration'].str.replace(' min', '')
    # Convertendo a duração para numérica apenas para filmes
    data.loc[data['type'] == 'Movie', 'duration'] = pd.to_numeric(data.loc[data['type'] == 'Movie', 'duration'], errors='coerce').astype('Int64')
    # Calculando a média de duração dos filmes
    media_duracao = data[data['type'] == 'Movie']['duration'].mean()
    print("Duração média dos filmes:", media_duracao)

# Avaliação média por gênero
# Se a coluna 'rating' estiver presente, calcular a média de avaliações por gênero
if 'rating' in data.columns:
    # Convertendo a coluna de ratings para numérica
    data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
    # Calculando a média de avaliações agrupadas por gênero
    media_avaliacao_por_genero = data.explode('listed_in').groupby('listed_in')['rating'].mean().sort_values(ascending=False)
    print("Avaliação média por gênero:")
    display(media_avaliacao_por_genero.head(10))

# PERGUNTAS:

# 1. Quais colunas estão presentes no dataset?
# Resposta: show_id, type, title, director, cast, country, date_added, release_year, rating, duration, listed_in, description.

# 2. Quantos filmes estão disponíveis na Netflix?
# Resposta: 6131

# 3. Quem são os 5 diretores com mais filmes e séries na plataforma?
# Resposta: Rajiv Chilaka (9); Raúl Campos, Jan Suter (18); Marcus Raboy (16); Suhas Kadav (16); Jay Karas (14).

# 4. Quais diretores também atuaram como atores em suas próprias produções?
# Resposta: Steven Spielberg, Antoine Fuqua, David de Vos, Nagesh Kukunoor, Spike Lee, Jim Henson, Richard Linklater, David Oyelowo, Dustin Hoffman, Rohit Shetty, Justin Baldoni, Jorge Blanco, Clint Eastwood, Trey Parker, Garry Marshall, Quentin Tarantino, Augustine Frizzell, Sofia Coppola, Tommy Chong, Alessandra de Rossi, Barry Levinson, Pascal Atuma, Udoka Oyeka, Youssef Chahine, Rano Karno, Ramsey Nouah, Malik Nejer, Mary Lambert, J.J. Abrams, Lynn Shelton, Vincent Ward, Nibal Arakji, Kimmy Gatewood, Anurin Nwunembom, Paul Weitz, Max Jabs, Myriam Fares, Don Michael Paul, Rob Reiner, Scott Spiegel, Michael Jai White, Alan Alda, Barbra Streisand, Bo Burnham, Yılmaz Erdoğan, Jennifer Brea, Sean McNamara, Vondie Curtis-Hall, Paul Greengrass, Justin Kelly, Kevin Smith, Seema Pahwa, Edward James Olmos, Ashutosh Gowariker, Aamir Khan, Guillermo del Toro, Ron Howard, TT The Artist, Angga Dwimas Sasongko, Kagiso Lediga, Chris Smith, Tate Taylor, Oliver Stone, Muharrem Gülmez, Mahsun Kırmızıgül, Amy Poehler, Sunny Lau, Kevin Costner, Matt Ross, Hamisha Daryani Ahuja, Ava DuVernay, Martin Scorsese, Jon Favreau, RZA, Peter Facinelli, Renuka Shahane, James Toback, Chris Rock, Sohail Khan, Robert Townsend, Sean Penn, Sabrina Rochelle Kalangie, Maïwenn, George Clooney, Muhammad Ali, John Putch, George C. Wolfe, Vir Das, Jalil Lespert, Detlev Buck, Phil Burgers, Brian Henson, Mohammad Adel, Debbie Allen, Anurag Kashyap, Siddique, Kunle Afolayan, Kevin Pollak, Otoja Abit, Peter Berg, Antonio Díaz, Natasha Lyonne, Glen Keane, Philippe Aractingi, Rana Eid, Neeraj Vora, Radha Blank, Christopher Nolen, Sudhir Mishra, Taika Waititi, Joe Mantello, Rajat Kapoor, Imtiaz Ali, Shawn Levy, Smeep Kang, Tigmanshu Dhulia, Chandra Liow, Satish Rajwade, Mahesh Manjrekar, Vijay Maurya, Tolga Karaçelik, Chris Miller, Isabel Sandoval, Steven Rinella, Mike Doyle, Peter Sullivan, Axelle Laffont, Funke Akindele, Nishikant Kamat, Kabir Sadanand, Umesh Shukla, Cem Yılmaz, Daniel Effiong, Rachel Griffiths, Musthafa, Manatsanun Phanlerdwongsakul, Kanika Batra, Michael Cristofer, Greta Gerwig, James Franco, Michael Showalter, Fouad El-Mohandes, He Xiaofeng, Lucas Margutti, Mysskin, Omoni Oboli, Sermiyan Midyat, Ahmed Saleh, Kris Pearn, Müfit Can Saçıntı, M. Night Shyamalan, Mahesh Bhatt, Karan Johar, Stephanie Turner, Toyin Abraham, Cristi Puiu, Angelina Jolie, Robert Krantz, Shannon Hartman, David Lynch, Tyler Perry, Numa Perrier, Tom Green, Ravi Babu, Emir Kusturica, Odunlade Adekola, Falz, Mark Lewis, Farhan Akhtar, Biodun Stephen, Trey Edward Shults, Mati Diop, Mike Ezuruonye, Eli Roth, Sergio Pablos, Neal Brennan, Parthiban, Michael A. Nickles, Daniel Sánchez Arévalo, Gupse Özay, Deepti Naval, Scott Aukerman, Jesse Moss, Camille Shooshani, Vijay Kumar, Jerry Seinfeld, Spike Jonze, Sanjay Jadhav, Hepi Mita, Yen Cheng-kuo, Syamsul Yusof, Alan Rickman, Beyoncé Knowles-Carter, Brie Larson, Amy Schumer, Chiwetel Ejiofor, Sridhar Rangayan, Michael Blieden, Parambrata Chatterjee, Kunal Kohli, Genevieve Nnaji, Ram, Kheiron, Rarecho, David Cronenberg, Alfonso Cuarón, Andy Serkis, Nam Ron, Dhanush, Nandita Das, Raditya Dika, Christopher Guest, Vincent Kok, Orson Welles, Rana Ranbir, Gaurav Narayanan, Akin Omotoso, Chris Eigeman, Eric Idle, Terry Jones, Deep Joshi, Shashanka Ghosh, Vijay, Sidharth Malhotra, Peter Ho, Samuel Jouy, Han Qing, Lauren Miller Rogen, Prabhu Deva, Samir Soni, Felix Starck, Tig Notaro, Bobcat Goldthwait, Jennifer Morrison, Amol Palekar, Daryl Hannah, Yoo Byung-jae, Anand Tiwari, Selima Taibi, David Wain, Jay Baruchel, Noël Wells, Patrick Brice, Umesh Mehra, Shreyas Talpade, Judah Friedlander, Errol Morris, Griffin Dunne, Reem Kherici, Mahmoud al Massad, Shammi Kapoor, Bryan Fogel, Maz Jobrani, Harry Chaskin, Stefan Brogren, Pat Healy, Lekh Tandon, Arun Chidambaram, Cal Seville, Jayaprakash Radhakrishnan, Jeff Garlin, Lucien Jean-Baptiste, Louis C.K., Eduardo Casanova, Lonny Price, Macon Blair, Raúl Arévalo, Ricky Gervais, Chester Tam, Rohit Mittal, Diego Luna, Ralph Macchio, Linas Phillips, Werner Herzog, Yvan Attal, Clovis Cornillac, Guilherme Fontes, Aamir Bashir, Christopher Martin, Cary Joji Fukunaga, Aziz Ansari, David Sampliner, Wyatt Cenac, Leif Tilden, Sam Upton, Trent Haaga, Tom Fassaert, Mark Williams, Chris Bell, Natalie Portman, Martin Lawrence, Kyle Rideout, Castille Landon, J. Michael Long, Vincent Perez, Jerry G. Angelo, Tim Blake Nelson, Sarah Smith, Justin Lee, Zoe Lister-Jones, Mike Judge, Joey Kern, Scott Martin, Nick Broomfield, Marianna Palka, Ethan Hawke, Raj Kapoor, Alê Abreu, Simon Baker, David McCracken, Brian De Palma, Wong Jing, Anuranjan Premji, Tom Holland, Corbin Bernsen, Eric Stoltz, Demetri Martin, Christopher Smith, Chia-Liang Liu, Todd Standing, Charles Martin Smith, Mike Birbiglia, Valeria Golino, Nagraj Manjule, Terry Gilliam, Luke Jurevicius, James Wong, Asri Bendacha, Justin Chon, Mathieu Kassovitz, Ananth Narayan Mahadevan, Satish Kaushik, Ilya Naishuller, Kelly Noonan, Corey Yuen, Note Chern-Yim, Akiva Schaffer, Ken Marino, Drew Casson, Chuck Russell, Sachin, Tony Leondis, Suman Mukhopadhyay, Jonathan Baker, Brad Bird, Eugenio Derbez, Dustin Nguyen, Kranti Redkar, Tinnu Anand.

# INSIGHTS ADICIONAIS:

# 1. Crescimento de Conteúdo ao Longo dos Anos

# Um dos insights mais significativos é o aumento no número de produções lançadas ao longo dos anos.
# Por exemplo, em 2017, houve 1.032 lançamentos, enquanto em 2021 esse número foi reduzido para 592.
# Isso pode indicar uma variação na estratégia da Netflix em relação à produção de conteúdo,
# possivelmente por mudanças na demanda ou em resposta à pandemia.

# 2. Gêneros mais Populares

# Os gêneros mais representados no catálogo da Netflix incluem:
# - International Movies (2.752 produções)
# - Dramas (2.427 produções)
# - Comedies (1.674 produções)
# Isso sugere que a Netflix está investindo fortemente em produções internacionais e dramáticas,
# possivelmente para atender a uma audiência global diversificada.

# 3. Diretores Prolíficos

# Os cinco diretores com mais filmes e séries na plataforma são:
# - Rajiv Chilaka: 19 produções
# - Raúl Campos, Jan Suter: 18 produções
# - Marcus Raboy: 16 produções
# - Suhas Kadav: 16 produções
# - Jay Karas: 14 produções
# Esses dados podem ser úteis para entender quais criadores estão mais ativos na plataforma
# e podem influenciar decisões de curadoria ou marketing.

# 4. Diretores que Também São Atores

# A lista de diretores que atuaram em suas próprias produções é extensa,
# incluindo nomes como Steven Spielberg, Spike Lee e Quentin Tarantino.
# Isso pode indicar uma tendência em que criadores multifacetados estão cada vez mais
# envolvidos em suas obras, o que pode enriquecer o conteúdo.

# 5. Duração Média dos Filmes
# A duração média dos filmes é de aproximadamente 99,6 minutos.
# Esse dado pode ser relevante para entender as preferências do público
# e as estratégias de produção, especialmente considerando que a maioria dos filmes tende a durar entre 90 e 120 minutos.

# 6. Medida das avaliações dos filmes
# Um ponto interessante é a média de avaliações, que aparece como NaN para muitos gêneros.
# Isso sugere que as avaliações podem não estar sendo registradas de maneira consistente
# ou que muitos filmes têm poucos ratings. Isso pode impactar as decisões dos espectadores
# ao escolher o que assistir, assim como a percepção da qualidade do conteúdo.

# Conclusão
# Esses insights podem ajudar a informar decisões estratégicas tanto para a Netflix
# quanto para criadores de conteúdo e analistas de mercado. A compreensão do que funciona
# e do que não funciona pode levar a melhores práticas de produção e a uma melhor experiência do usuário na plataforma.

# By Daniel Tavares
```
